### 문제 
- health.csv에서 마지막 열이 분류 값이다. 마지막열을 제외한 나머지를 이용하여 분류값을 예측하는 프로그램을 작성하시오.
- 종속변수는 1,2,3 중 하나
- 2가지 방법(Random Forest, SVC)를 이용하여 예측하되, test data에 대한 accuracy 가 최대가 되도록 하시오.
- training data 와 test data의 비율이 7:3 이 되도록 하고, random_state = 2022 를 사용하세요.

- 각 방법별로 test data에 대한 accuracy를 인쇄하시오.
- 맨 마지막에 두 방법의 test accuracy중 최대값을 인쇄하시오.

### 채점: 최대 test accuracy 순으로 정렬하여 1등 30점부터 등수마다 2점씩 차감하여, 15등 이하는 2점

#### 최종 학점 tie-break 룰: 합계 점수가 동점인 경우 기말 점수가 높은 수강생에게 높은 학점을 배정
