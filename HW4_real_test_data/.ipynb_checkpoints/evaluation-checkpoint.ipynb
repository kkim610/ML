```python
from pickle import load

import os.path
from os import path
import pandas as pd

from sklearn.metrics import f1_score
```

```python
X_test=pd.read_csv("test.csv",header=None).to_numpy()
```

```python
y_test=X_test[:,5]
```

```python
X_test2=X_test[:,:-1]
```

```python
f1_L=[]
```

```python
# load the model

if path.exists("svm_model.pkl"):
    svm_model = load(open('svm_model.pkl', 'rb'))
else:
    print("svm_model.pkl does not exist !!!!")

# load the scaler
if path.exists("svm_scaler.pkl"):
    scaler = load(open('svm_scaler.pkl', 'rb'))
    X_test = scaler.transform(X_test2)
else:
    X_test = X_test2.copy()

# make predictions on the test set
y_pred = svm_model.predict(X_test)

# calculate f1-score
f1_L.append(f1_score(y_test, y_pred, average='micro'))
f1_L
```
